```python
# 1. Write a function that decides for given a, b ∈ N+ numbers whether a is a divisor of b. Solve
# it in at least three ways. Use the built-in function a.divides(b) only for testing: check the
# solution for all integers a and b between 1 and 100.

def is_divisor(a: int, b: int) -> bool:
    return b % a == 0

def is_divisor2(a: int, b: int) -> bool:
    while a < b:
        a += a
    return False

def is_divisor3(a: int, b: int) -> bool:
    while a <= b:
        if a == b:
            return True
        a *= 2
    return False
```

```python
# 2. Write a function that returns the list of positive divisors of a number. Don’t use the built-in
# function divisors() other than for testing.

from math import ceil


def pos_divs(n: int)->list[int]:
    _r:  list[int] = []
    for i in range(1, ceil(n/2)):
        if (_r % i == 0):
            _r.append(i)
    return _r   
```

```python
# 4. Write a function that decides whether a given n ∈ N is prime. Don’t use the built-in number
# theory functions of Sage. Check the solution using the built-in function is_prime() from 0 to
# 1000

def is_prime_(n: int)-> bool:
  if (n < 2): 
    return False
  for i in range(2, n):
    if (n % i == 0):
      return False
  return True
```

```python
all([is_prime_(x) == is_prime(x) for x in range(0,1000)])
```

```python
# 9. Implement the sieve of Eratosthenes: write a function that returns all prime numbers in a list
# up to a given n. (Don’t use the built-in prime-related functions.)

def sieve(n: int) -> List[int]:
  _r: List[int] = [i for i in range(2, n + 1)]
  
  for i, n1 in enumerate( _r):
    if n1 == 0:
      continue
    for j, n2 in enumerate(_r):
      if (j > i and n2 % n1 == 0):
        _r[j] = 0
  return filter(lambda x: x > 0, _r)
```

```python
len(list(sieve(100)))
```
